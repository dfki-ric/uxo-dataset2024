## Real data

```python
from common.config import get_config
config = get_config()
```

```python
ROOT= '../' + config['aris_extract']
DATASET = '2023-09-20_171105'
FILENAME_ARIS_METADATA = 'aris_frame_meta.csv'
FILENAME_CRANE = 'gantry.csv'
```

```python
df_data_aris = pd.read_csv(os.path.join(ROOT, DATASET, FILENAME_ARIS_METADATA))
df_pos = pd.read_csv(os.path.join(ROOT, DATASET, FILENAME_CRANE))
df_pos
```

```python
df_data_clean = df_pos
df_data_clean[['roll', 'pitch', 'yaw']] = df_data_aris[['SonarRoll', 'SonarTilt', 'SonarPan']]
df_data_clean
```

```python
poss = []
for index, row in df_data_clean.iterrows():
    A2B = np.eye(4) 
    A2B[:3,3] = np.array(row[['x','y','z']])
    tm.add_transform("setup/portal_crane", "world", A2B=A2B)
    poss.append(tm.get_transform("setup/ar3", "world").round(6)[:3, 3])

```

```python
from scipy.spatial.transform import Rotation as R
rots = [(R.from_matrix(tm.get_transform("setup/ar3", "world")[:3,:3]) * R.from_euler('xyz', row[['roll','pitch','yaw']], degrees=True)).as_quat() for index, row in df_data_clean.iterrows()]
```

```python
df_ar3 = pd.DataFrame()
df_ar3['aris_frame_idx'] = df_data_clean['aris_frame_idx']
df_ar3[['pos.x','pos.y', 'pos.z']] = poss
df_ar3[['rot.x','rot.y', 'rot.z', 'rot.w']] = rots
df_ar3
```

```python
df_ar3.to_csv(os.path.join(ROOT, DATASET, 'ar3.csv'), index=False)
```

```python
from pytransform3d.plot_utils import make_3d_axis
import pytransform3d.transformations as pytr
from pytransform3d import visualizer as pv
```

```python
ROOT_MESH = os.path.join('..', '..', '3d_models')
SCALING=20

fig = pv.figure()
fig.plot_mesh(filename=os.path.join(ROOT_MESH, "100lbs_aircraft_bomb/100lbs_aircraft_bomb.stl"), s=np.ones(3), A2B=tm.get_transform("target/100lbs", "world"))

fig.plot_transform(A2B=tm.get_transform("target/100lbs", "world"), s=0.2)

for index, row in df_ar3.iloc[::SCALING].iterrows():

    tm.add_transform("setup/ar3", "world", A2B=pytr.transform_from(R.from_quat(row[['rot.x','rot.y', 'rot.z', 'rot.w']]).as_matrix(), row[['pos.x','pos.y', 'pos.z']]))
    fig.plot_transform(A2B=tm.get_transform("setup/camera", "world"), s=0.2)

fig.show()
```

```python
from pytransform3d.plot_utils import make_3d_axis
import pytransform3d.transformations as pytr

SCALING = 50
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection="3d")
ax = make_3d_axis(ax_s=1, unit="m", n_ticks=6)
for index, row in df_ar3.iloc[::SCALING].iterrows():
    tm.add_transform("setup/ar3", "world", A2B=pytr.transform_from(R.from_quat(row[['rot.x','rot.y', 'rot.z', 'rot.w']]).as_matrix(), row[['pos.x','pos.y', 'pos.z']]))
    pytr.plot_transform(ax, A2B=tm.get_transform("setup/camera", "world"), s=0.2)
    pytr.plot_transform(ax, A2B=tm.get_transform("setup/ar3", "world"), s=0.2)

pytr.plot_transform(ax, A2B=tm.get_transform("target/100lbs", "world"), s=0.2)
pytr.plot_transform(ax, A2B=tm.get_transform("world", "world"), s=0.2)



plt.show()
```
